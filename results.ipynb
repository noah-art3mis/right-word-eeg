```python
import pandas as pd
import pingouin as pg
import matplotlib.pyplot as plt
import pingouin as pg

df = pd.read_csv("data/processed/alpha_beta.csv")
df
```

```python
df_alpha = df[
    [
        "neo_alpha",
        "tech_alpha",
        "modern_alpha",
        "brutal_alpha",
        "bio_alpha",
        "park_alpha",
    ]
]

df_alpha = df_alpha.rename(
    columns={
        "neo_alpha": "neoclassical",
        "tech_alpha": "high_tech",
        "modern_alpha": "modernist",
        "brutal_alpha": "brutalist",
        "bio_alpha": "biophilic",
        "park_alpha": "park",
    }
)

df_alpha
```

```python
df_beta = df[
    [
        "neo_beta",
        "tech_beta",
        "modern_beta",
        "brutal_beta",
        "bio_beta",
        "park_beta",
    ]
]

df_beta = df_beta.rename(
    columns={
        "neo_beta": "neoclassical",
        "tech_beta": "high_tech",
        "modern_beta": "modernist",
        "brutal_beta": "brutalist",
        "bio_beta": "biophilic",
        "park_beta": "park",
    }
)

df_beta
```

## Check normality


```python
import scipy.stats as stats
import seaborn as sns

fig, axes = plt.subplots(nrows=2, ncols=6, figsize=(20, 8))

for i, col in enumerate(df_beta.columns):
    row, col_idx = divmod(i, 6)

    sns.histplot(df_beta[col], kde=True, ax=axes[0, col_idx])
    axes[0, col_idx].set_title(f"Histogram: {col}")

    pg.qqplot(df_beta[col], dist="norm", ax=axes[1, col_idx])
    axes[1, col_idx].set_title(f"Q-Q Plot: {col}")

plt.tight_layout()
plt.show()
```

```python
import scipy.stats as stats
import seaborn as sns

fig, axes = plt.subplots(nrows=2, ncols=6, figsize=(20, 8))

for i, col in enumerate(df_alpha.columns):
    row, col_idx = divmod(i, 6)

    sns.histplot(df_alpha[col], kde=True, ax=axes[0, col_idx])
    axes[0, col_idx].set_title(f"Histogram: {col}")

    pg.qqplot(df_alpha[col], dist="norm", ax=axes[1, col_idx])
    axes[1, col_idx].set_title(f"Q-Q Plot: {col}")

plt.tight_layout()
plt.show()
```

Data seems reasonably normal and was collected in independent samples, so we use a one-way ANOVA.


```python
df_alpha = df_alpha.melt(var_name="Style", value_name="Value")
df_beta = df_beta.melt(var_name="Style", value_name="Value")
```

```python
# alphas
sns.violinplot(data=df_alpha, x="Style", y="Value")
```

```python
pg.anova(data=df_alpha, dv="Value", between="Style")
```

```python
# betas
sns.violinplot(data=df_beta, x="Style", y="Value")
```

```python
pg.anova(data=df_beta, dv="Value", between="Style")
```
